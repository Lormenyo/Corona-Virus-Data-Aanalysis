```python
import csv
```

```python
cases = []
population = []


with open('covid_data.csv') as csvfile:
    f = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in f:
         cases.append(row)
         
```

```python
with open('population_data.csv') as csvfile:
    p = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in p:
        population.append(row)
            
population
```

```python
cases
```

```python
cases_dict = {}
deaths_dict = {}
for row in cases[1:]:
    country = row[1]
    if country in cases_dict.keys():
        cases_dict[country] += int(row[2])
        deaths_dict[country] += int(row[3])
    else:
        cases_dict[country] = int(row[2])
        deaths_dict[country] = int(row[3])
```

```python
infection_values = list(cases_dict.values())
sorted_values = list(cases_dict.values())
sorted_values.sort(reverse=True)
sorted_values[:10]
```

```python
# Country with the highest infection
highestInfectionCountry = list(cases_dict.keys())[infection_values.index(sorted_values[0])]
#number of infections in that country
highestInfection = sorted_values[0]
print(highestInfectionCountry, highestInfection)
```

```python
# Country with the second highest infection
secondHighestInfectionCountry = list(cases_dict.keys())[infection_values.index(sorted_values[1])]
#number of infections in that country
secondHighestInfection = sorted_values[1]
print(secondHighestInfectionCountry, secondHighestInfection)
```

```python
population_dict = {}
for row in population[1:]:
    country = row[0]
    if len(row) > 4:
        row.remove(row[2])
        population_dict[country] = int(row[2])
    else:
        population_dict[country] = int(row[2])
        
population_dict
```

```python
#Highest Infection rate(ratio of the number of infections to population)
infectionRates = {}
for country in population_dict.keys():
    if country not in cases_dict.keys():
        pass
    else:
        infectionRate = cases_dict[country]/population_dict[country]
        infectionRates[country] = infectionRate
        
infectionRates
```

```python
infectionRate_values = list(infectionRates.values())
sortedRates_values = list(infectionRates.values())
sortedRates_values.sort(reverse=True)
sortedRates_values[:10]
```

```python
# Country with the Highest Infection rate(ratio of the number of infections to population)
highestInfectionRateCountry = list(infectionRates.keys())[infectionRate_values.index(sortedRates_values[0])]
#Infection Rate in that country
highestInfectionRate = sortedRates_values[0]
print(highestInfectionRateCountry, highestInfectionRate)
```

```python
#overall death rate (ratio of number of deaths to number of infections) 
overallDeathRate = sum(list(deaths_dict.values()))/sum(infection_values)
overallDeathRate
```

```python
# country that has the highest death rate (ratio of number of deaths to number of infections)
# What is the death rate for that country?
deathRates = {}
for country in cases_dict.keys():
    deathRate = deaths_dict[country]/cases_dict[country]
    deathRates[country] = deathRate
        
deathRates
```

```python
deathRate_values = list(deathRates.values())
sortedDeathRates = list(deathRates.values())
sortedDeathRates.sort(reverse=True)
sortedDeathRates[:10]

# Country with the Highest Death rate (ratio of number of deaths to number of infections)
highestDeathRateCountry = list(deathRates.keys())[deathRate_values.index(sortedDeathRates[0])]
#Infection Rate in that country
highestDeathRate = sortedDeathRates[0]
print(highestDeathRateCountry, highestDeathRate)
```

```python
#  number of new infections per day on the rise (i.e. a positive trend over the most recent 1 week of data)
import datetime
timeSeriesFile = open('partial_time_series.csv', 'r')
timeSeries = [r.strip("\r\n") for r in timeSeriesFile.readlines()]
timeSeries 
windowlength = len(timeSeries)
date_dict = {}
for row in cases[1:]:
    date = row[0]
    if date not in date_dict.keys():
        date_dict[date] = [(row[1], row[2])]
     
    else:
        date_dict[date].append((row[1], row[2]))
dates = list(date_dict.keys())
dates.sort(key=lambda date: datetime.datetime.strptime(date, "%m/%d/%y"))

infectionPerDay    = {}
# reversedCases = cases[1:]
# reversedCases.reverse()
for row in cases[1:]:
    country = row[1]
    if country not in infectionPerDay.keys():
        infectionPerDay[country] = [int(row[2])/6,1]
    elif country in infectionPerDay.keys() and infectionPerDay[country][1] !=7:
        new_slope = infectionPerDay[country][0] - int(row[2])/6
        infectionPerDay[country][0] += new_slope
        infectionPerDay[country][1] += 1
    else:
        pass
    
    if row[0] == dates[0]:
        
        
negativeInfectionCountries = []
positiveInfectionCountries = []
maxIncrease = list(infectionPerDay.values())[0][0]
maxDecrease = list(infectionPerDay.values())[0][0]
for dailyInfection in list(infectionPerDay.values()):
#     dailyInfectionValues.append(dailyInfection[0])
    country = list(infectionPerDay.keys())[list(infectionPerDay.values()).index(dailyInfection)]
    if dailyInfection[0] > 0.0:
        positiveInfectionCountries.append(country)
    elif dailyInfection[0] == 0.0:
        pass
    else:
        negativeInfectionCountries.append(country)
        
    
    if dailyInfection[0] < 0.0 and dailyInfection[0] < float(maxDecrease):
        maxDecreaseCountry, maxDecrease = country, dailyInfection[0]

    elif dailyInfection[0] > 0.0 and dailyInfection[0] > float(maxIncrease):
        maxIncreaseCountry, maxIncrease = country, dailyInfection[0]
    else:
        pass
 

# print(reversedCases)
print("Negative Infection per Day ")
print(negativeInfectionCountries)
print("Positive Infection per Day ")
print(positiveInfectionCountries)
print("Max Increase = ", maxIncrease, maxIncreaseCountry)
print("Max Decrease = ", maxDecrease, maxDecreaseCountry)
```
